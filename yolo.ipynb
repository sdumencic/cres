# Prediction

Predicts the bounding boxes and saves them to the same json format that is exported by validation in batches of 20 images

```python
from ultralytics import YOLO
import glob
import json
import os
from datetime import datetime

now = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
results_dir = f"./results_{now}"
os.makedirs(results_dir, exist_ok=True)

model_path = f'./best.pt'
model = YOLO(model_path)

# Run batched inference on a list of images
image_paths = glob.glob(r"D:\DCIM\100MEDIA\*.JPG") # D:\DCIM\100MEDIA\*.JPG
batch_size = 20
predictions = []

for batch_start in range(0, len(image_paths), batch_size):
    batch_paths = image_paths[batch_start:batch_start + batch_size]
    batch_results = model(batch_paths)
    batch_predictions = []
    for result in batch_results:
        boxes = result.boxes
        if boxes is not None and boxes.xyxy is not None:
            image_path = result.path
            image_id = os.path.splitext(os.path.basename(image_path))[0]
            xywh = boxes.xywh.cpu().numpy()
            conf = boxes.conf.cpu().numpy()
            cls = boxes.cls.cpu().numpy().astype(int)
            for j in range(len(xywh)):
                bbox = xywh[j].tolist()
                prediction = {
                    "image_id": image_id,
                    "category_id": int(cls[j]),
                    "bbox": bbox,
                    "score": float(conf[j]),
                    "timestampDetected": datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
                }
                batch_predictions.append(prediction)
                predictions.append(prediction)
                
    batch_json_path = f"{results_dir}/pred_bboxes_batch_{batch_start//batch_size + 1}.json"
    with open(batch_json_path, "w") as f:
        json.dump(batch_predictions, f, indent=2)
```

# Combining json files

Combines all json file batches to a single json file

```python
import glob
import json

batch_json_files = sorted(glob.glob(f"{results_dir}/pred_bboxes_batch_*.json"))

all_predictions = []
for file in batch_json_files:
    with open(file, "r") as f:
        batch_preds = json.load(f)
        all_predictions.extend(batch_preds)

with open("predictions_combined.json", "w") as f:
    json.dump(all_predictions, f, indent=2)

print(f"Combined {len(batch_json_files)} batch files into pred_bboxes.json with {len(all_predictions)} predictions.")
```
